```python
import pandas as pd
import mysql.connector

# Connecting to MySQL database using mysql-connector-python
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='WJ28@krhps',
    database='superstoresystem'
)

# Function to fetch data into a DataFrame
def fetch_data(query):
    cursor = connection.cursor()
    cursor.execute(query)
    rows = cursor.fetchall()
    colnames = [desc[0] for desc in cursor.description]
    df = pd.DataFrame(rows, columns=colnames)
    cursor.close()
    return df

# Query to select suppliers' SupplierID and SupplierName in ascending order
Suppliers_SupplierID_query_asc = """
SELECT SupplierID, SupplierName
FROM Suppliers
ORDER BY SupplierID ASC;
"""

# Fetching data and storing it in a DataFrame
Suppliers_SupplierID_df_asc = fetch_data(Suppliers_SupplierID_query_asc)

print("Suppliers SupplierID in Ascending Order:")
print(Suppliers_SupplierID_df_asc)

# Close the database connection
connection.close()
```
